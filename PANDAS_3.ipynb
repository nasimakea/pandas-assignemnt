# Q6. What do you understand about the windows function in pandas and list the types of windows functions?

# ANS


#"""
We, as users of pandas, understand that window functions in pandas are a powerful feature for performing operations on a set of data points within a specified window or range. These functions are particularly useful for time-series data and other scenarios where analysis within a moving or expanding window is required.

Pandas provides several types of window functions, allowing us to perform various operations within the specified window. Some of the common types of window functions include:

Rolling Functions:

These functions operate on a specified window of fixed size that "rolls" or moves over the data.
Examples include rolling.mean(), rolling.sum(), and rolling.std().
Expanding Functions:

These functions consider all the data points from the beginning up to the current point.
Examples include expanding.mean(), expanding.sum(), and expanding.min().
Exponential Moving Average (EMA):

EMA is a type of weighted moving average where more recent data points are given higher weight.
It can be calculated using the ewm() function in pandas.
Aggregate Functions:

These functions perform operations on data points within a window and aggregate the results into a single value.
Examples include agg(), which allows custom aggregation functions to be applied.
Windowed Groupby Operations:

Pandas supports window operations on grouped data using the groupby() function along with window functions."""

#Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
#calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

# ANS:-

```python
def calculate_date_difference(start_date, end_date):

    start_datetime = pd.to_datetime(start_date)
    end_datetime = pd.to_datetime(end_date)


    time_difference = end_datetime - start_datetime

    days = time_difference.days
    hours, remainder = divmod(time_difference.seconds, 3600)
    minutes, _ = divmod(remainder, 60)

    return days, hours, minutes
```

```python
print(calculate_date_difference('2024-01-1', '2024-01-26'))
```
